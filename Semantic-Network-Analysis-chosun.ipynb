## Module Import

```python
# self defined Modules
from myModules.plot.plotNetwork import Network

# General Modules
import warnings
import pickle

warnings.filterwarnings('ignore')
```

## Data Load

```python
DATA_ROOT = './processed-data/chosun/'

RESULT_ROOT = './Result/chosun/'
```

```python
with open(DATA_ROOT+"lemmatized-noun.pkl", "rb") as f:
    noun = pickle.load(f)
with open(DATA_ROOT+"lemmatized-verb.pkl", "rb") as f:
    verb = pickle.load(f)
with open(DATA_ROOT+"lemmatized-adjective.pkl", "rb") as f:
    adjective = pickle.load(f)
with open(DATA_ROOT+"lemmatized-adverb.pkl", "rb") as f:
    adverb = pickle.load(f)
```

## 3. Network Analysis

### 3-1. Noun

```python
network = Network(data=noun, num_node=30)
```

```python
network.plot(centrality='degree', title='Noun-Network_Analysis', root=RESULT_ROOT)
```

```python
network.plot(centrality='closeness', title='Noun-Network_Analysis', root=RESULT_ROOT)
```

```python
network.plot(centrality='betweenness', title='Noun-Network_Analysis', root=RESULT_ROOT)
```

```python
network.plot(centrality='eigenvector', title='Noun-Network_Analysis', root=RESULT_ROOT)
```

```python
df = network.save_centrality_table(title='noun-centrality-table', root=RESULT_ROOT)
df
```
